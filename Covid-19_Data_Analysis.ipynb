```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('covid_india.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

# Active Cases Across States 

```python
sns.set(rc={'figure.figsize':(15,5)})
ax= sns.barplot(data=data,x='Name of State / UT', y='Active Cases')
ax.set_title('Active Cases by State')
plt.xticks(rotation=90)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

*The graph indicates that Kerala, Maharashtra, and Karnataka have the highest number of active cases*

# Cured/Discharged/Migrated Across States

```python
sns.set(rc={'figure.figsize':(15,5)})
ax= sns.barplot(data=data,x='Name of State / UT', y='Cured/Discharged/Migrated')
ax.set_title('Cured rate by State')
plt.xticks(rotation=90)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

*This will visually emphasize Maharashtra, Kerala, and Karnataka as the top states with the highest recovered cases*

# Deaths Across States

```python
ax= sns.barplot(data=data,x='Name of State / UT', y='Deaths')
ax.set_title('Cured rate by State')

plt.xticks(rotation=90)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

*This will visually emphasize Maharashtra, Karnataka and Tamil Nadu as the top states with the highest Death cases*

# Case fatality rate (CFR)

```python
plt.figure(figsize=(10, 10))
plt.pie(data['Deaths'], labels=data['Name of State / UT'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"), labeldistance=1.1 )

plt.xticks(rotation=45)

# Add a title
plt.title('Deaths as % of Total Cases by State', fontsize=16)

# Show the plot
plt.show()
```

*This will visually emphasize  Maharashtra, Karnataka and Tamil Nadu as the top states with the highest Case fatality rate (CFR)s by State*

# Conclusion :

Based on the data analysis, the following key insights can be drawn:  

1. **Active Cases**:  
   - Kerala, Maharashtra, and Karnataka have the highest number of active cases.  
   - This suggests that these states are currently experiencing a significant burden of ongoing infections, which may require increased medical resources and containment measures.  

2. **Recovered Cases**:  
   - Maharashtra, Kerala, and Karnataka have the highest number of recovered (Cured/Discharged/Migrated) cases.  
   - This indicates that while these states have faced a high number of infections, they have also successfully managed recoveries, likely due to better healthcare infrastructure and patient management.  

3. **Deaths**:  
   - Maharashtra, Karnataka, and Tamil Nadu have reported the highest number of deaths.  
   - This highlights the severe impact of the pandemic in these states, possibly due to a combination of factors such as population density, healthcare accessibility, and comorbidities among affected individuals.  

4. **Case fatality rate (CFR)**:  
   - Maharashtra, Karnataka, and Tamil Nadu also have the highest death rates relative to their total cases.  
   - This suggests that these states may have higher fatality rates, which could be due to delayed hospitalization, inadequate critical care capacity, or higher vulnerability among the infected population.  

#  **Final Insights**  
- **Maharashtra and Karnataka appear in all four categories**, indicating that they have been among the worst-hit states in terms of active cases, recoveries, and fatalities.  
- **Kerala has a high number of active and recovered cases but is not in the top death categories**, possibly suggesting a better survival rate or effective healthcare interventions.  
- **Tamil Nadu has a lower active case burden but ranks high in deaths and death percentage, indicating a potentially higher fatality rate among infected individuals.**  

Overall, these findings emphasize the need for **targeted healthcare strategies**, such as **boosting ICU capacity, improving early detection, and strengthening post-recovery care in high-risk states**.
